## Agregar datos por usuario

[Ahora que los datos est√°n limpios, agrega los datos por usuario y por periodo para que solo haya un registro por usuario y por periodo. Esto facilitar√° mucho el an√°lisis posterior.]

```python
# Calcula el n√∫mero de llamadas hechas por cada usuario al mes. Guarda el resultado.
numero_llamadas=llamadas.groupby([llamadas['user_id'], llamadas['call_date'].dt.month]).agg(numero_llamadas=('id','count')).reset_index()
numero_llamadas= numero_llamadas.rename(columns={'call_date':'mes_consumo'})
numero_llamadas
```

```python
# Calcula la cantidad de minutos usados por cada usuario al mes. Guarda el resultado.
cantidad_minutos = llamadas.groupby([llamadas['user_id'],llamadas['call_date'].dt.month]).agg(cantidad_minutos=('duracion_redondeada','sum')).reset_index()
cantidad_minutos = cantidad_minutos.rename(columns={'call_date':'mes_consumo'})
cantidad_minutos
```

```python
# Calcula el n√∫mero de mensajes enviados por cada usuario al mes. Guarda el resultado.
numero_mensajes = mensajes.groupby([mensajes['user_id'],mensajes['message_date'].dt.month]).agg(mensajes_enviados=('id','count')).reset_index()
numero_mensajes = numero_mensajes.rename(columns={'message_date': 'mes_consumo'})
numero_mensajes
```

```python
# Calcula el volumen del tr√°fico de Internet usado por cada usuario al mes. Guarda el resultado.
internet_usado = internet.groupby([internet['user_id'],internet['session_date'].dt.month]).agg(internet_usado=('internet_redondeado','sum')).reset_index()
internet_usado = internet_usado.rename(columns={'session_date':'mes_consumo'})
internet_usado
```

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumi√≥ un usuario √∫nico en un mes determinado.]

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month
fusion = numero_llamadas.merge(cantidad_minutos, on=['user_id','mes_consumo'],how='outer')
fusion = fusion.merge(numero_mensajes, on=['user_id','mes_consumo'],how='outer')
fusion = fusion.merge(internet_usado, on=['user_id','mes_consumo'], how='outer')
fusion
```

Hiciste una correcta agrupaci√≥n de los datos combinando `groupby`  y `merge`, pero recuerda que, seg√∫n el enunciado del proyecto, los valores de duraci√≥n de llamadas y consumo de internet deben redondearse hacia arriba, ya que el cobro se hace por minuto y GB completo. Para esto debes tener en cuenta lo siguiente:

* **Llamadas:** Primero realiza el redondeo hacia arriba de los minutos para cada llamada y luego sumas los minutos totales por usuario para cada mes.

* **Internet:** Primero sumas el consumo total en MB para cada usuario por mes, luego conviertes ese valor a GB y finalmente realizas el redondeo hacia arriba.

Los valores finales que obtienes para cada usuario, los comparas con los minutos y GB incluidos en cada plan, con el fin de establecer si hubo un exceso en el consumo y hacer el cobro respectivo.

```python
# A√±ade la informaci√≥n de la tarifa
fusion= pd.merge(fusion,usuarios[['user_id','plan']], on='user_id', how='left')
fusion = pd.merge(fusion,tarifas, left_on='plan', right_on='plan_name', how='left', suffixes=('', '_tarifa'))
fusion
```

```python
# Calcula el ingreso mensual para cada usuario
#Sustituimos valores ausentes por 0 ya que cero significa que no hubo consumo en esas fechas 
fusion['numero_llamadas'] = fusion['numero_llamadas'].fillna(0)
fusion['mensajes_enviados'] = fusion['mensajes_enviados'].fillna(0)
fusion['internet_usado'] = fusion['internet_usado'].fillna(0)
fusion['cantidad_minutos'] = fusion['cantidad_minutos'].fillna(0)

#Calculo de excesos (solo valores)
fusion['exceso_minutos'] = (fusion['cantidad_minutos'] - fusion['minutes_included']).clip(lower=0)
fusion['exceso_mensajes'] = (fusion['mensajes_enviados'] - fusion['messages_included']).clip(lower=0)
fusion['exceso_internet_gb'] = ((fusion['internet_usado'] - fusion['mb_per_month_included']) / 1024).clip(lower=0)

#Calculo de cargos por excesos
fusion['cargo_minutos_excedidos'] = fusion['exceso_minutos'] * fusion['usd_per_minute']
fusion['cargo_mensajes_excedidos'] = fusion['exceso_mensajes'] * fusion['usd_per_message']
fusion['cargo_internet_excedido'] = fusion['exceso_internet_gb'] * fusion['usd_per_gb']
fusion
```

## Prueba las hip√≥tesis estad√≠sticas

[Prueba la hip√≥tesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hip√≥tesis nula y alternativa, escoge la prueba estad√≠stica, determina el valor alfa.]

```python
# Prueba las hip√≥tesis
# Hip√≥tesis nula: Œº ‚â§ 40 (El ingreso promedio mensual es menor o igual a 40 USD)
# Hip√≥tesis alternativa: Œº > 40 (El ingreso promedio mensual es mayor a 40 USD)
# Valor de alfa =0,005 (5%)
# Calcular la media y varianza del ingreso_total
media_ingreso = fusion['ingreso_total'].mean()
varianza_ingreso = fusion['ingreso_total'].var()

print(f"Media del ingreso total: ${media_ingreso:.2f}")
print(f"Varianza del ingreso total: {varianza_ingreso:.2f}")
print(f"Desviaci√≥n est√°ndar del ingreso total: ${fusion['ingreso_total'].std():.2f}")

print("\nEstad√≠sticos descriptivos del ingreso total:")
print(fusion['ingreso_total'].describe())
from scipy import stats
muestra = fusion['ingreso_total']
n = len(muestra)  # tama√±o de la muestra
media_muestral = muestra.mean()
desviacion_std = muestra.std(ddof=1)  # desviaci√≥n est√°ndar muestral
valor_hipotesis = 40  # valor que quieres probar
alpha = 0.05  # nivel de significancia

# Calcular el estad√≠stico t
t_estadistico = (media_muestral - valor_hipotesis) / (desviacion_std / np.sqrt(n))

# M√©todo 1: Calcular el p-valor
p_valor = 1 - stats.t.cdf(t_estadistico, df=n-1)  # cola derecha

# M√©todo 2: Calcular el valor cr√≠tico
valor_critico = stats.t.ppf(1-alpha, df=n-1)  # para cola derecha

print(f"Estad√≠stico t: {t_estadistico:.4f}")
print(f"P-valor: {p_valor:.6f}")
print(f"Valor cr√≠tico: {valor_critico:.4f}")
print(f"Grados de libertad: {n-1}")
```

[Prueba la hip√≥tesis de que el ingreso promedio de los usuarios del √°rea NY-NJ es diferente al de los usuarios de otras regiones.]

[Elabora las hip√≥tesis nula y alternativa, escoge la prueba estad√≠stica, determina el valor alfa.]

```python
# Prueba las hip√≥tesis
# Agregar informaci√≥n de usuarios al DataFrame fusion
fusion_con_region = pd.merge(fusion, usuarios[['user_id', 'city']], on='user_id', how='left')

# Crear variable para identificar regi√≥n NY-NJ
fusion_con_region['es_ny_nj'] = fusion_con_region['city'].str.contains('New York-Newark-Jersey City', na=False)

# Separar los datos por regi√≥n
ingresos_ny_nj = fusion_con_region[fusion_con_region['es_ny_nj'] == True]['ingreso_total']
ingresos_otras = fusion_con_region[fusion_con_region['es_ny_nj'] == False]['ingreso_total']

print(f"Usuarios NY-NJ: {len(ingresos_ny_nj)}")
print(f"Usuarios otras regiones: {len(ingresos_otras)}")
print(f"Media NY-NJ: ${ingresos_ny_nj.mean():.2f}")
print(f"Media otras regiones: ${ingresos_otras.mean():.2f}")

t_estadistico, p_valor = stats.ttest_ind(ingresos_ny_nj, ingresos_otras)

print(f"Estad√≠stico t: {t_estadistico:.4f}")
print(f"P-valor: {p_valor:.6f}")
print(f"Nivel de significancia (Œ±): 0.05")

# Interpretaci√≥n
if p_valor < 0.05:
    print("\nDecisi√≥n: Rechazar H‚ÇÄ")
    print("Conclusi√≥n: Hay evidencia estad√≠stica de diferencia significativa entre regiones")
else:
    print("\nDecisi√≥n: No rechazar H‚ÇÄ")
    print("Conclusi√≥n: No hay evidencia estad√≠stica de diferencia significativa entre regiones")
```
